# Argentina Happiness Report
On this notebook I will be working with the Wold Happiness Report 2021. This job was inspired by https://www.kaggle.com/ajaypalsinghlo/world-happiness-report-2021-word, which is focused on India. I will focus this analysis on Argentina and the countries of the Mercosur.
The World Happiness Report is a publication of the Sustainable Development Solutions Network, powered by data from the Gallup World Poll and Lloydâ€™s Register Foundation, who provided access to the World Risk Poll.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file="world-happiness-report.csv"
world=pd.read_csv(file)
print(world.columns)
world.head()
```

On this first dataset we have information back up to 2005. The Happiness Report provides information about 6 topics: GDP per capita, social support, healthy life expectancy, freedom to make life choices, generoristy and perception of corruption. Let's focus on Argentina and let's see how all the variables changed through the years.

```python
argentina=world[world['Country name']=='Argentina']
argentina
```

```python
plt.figure(figsize=(25,8))
sns.lineplot(data=argentina, x="year", y="Life Ladder", markers=True)
plt.title('Life Ladder', fontsize=24)
```

```python
plt.figure(figsize=(25,8))
sns.lineplot(data=argentina, x='year', y='Log GDP per capita')
plt.title('Logged GDP per capita', fontsize=24)
```

```python
plt.figure(figsize=(25,8))
sns.lineplot(data=argentina, x='year', y='Social support')
plt.title('Social Support', fontsize=24)
```

```python
plt.figure(figsize=(25,8))
sns.lineplot(data=argentina, x='year', y='Healthy life expectancy at birth')
plt.title('Healthy Life Expectancy at birth', fontsize=24)
```

```python
plt.figure(figsize=(25,8))
sns.lineplot(data=argentina, x='year', y='Freedom to make life choices')
plt.title('Freedom to make life choices', fontsize=24)
```

```python
plt.figure(figsize=(25,8))
sns.lineplot(data=argentina, x='year', y='Generosity')
plt.title('Generosity', fontsize=24)
```

```python
plt.figure(figsize=(25,8))
sns.lineplot(data=argentina, x='year', y='Perceptions of corruption')
plt.title('Percepcions of Corruption', fontsize=24)
```

Now we'll work with a second dataset which contains data about the report made for 2021.

```python
report_2021=pd.read_csv('world-happiness-report-2021.csv')
print(report_2021.head())
```

Let's now focus on the countries that are part ot the Mercosur.
The Mercosur (Mercado ComÃºn del Sur or Southern Common Market) is a trade bloc founded in 1991. Its full members are Argentina, Brazil, Paraguay, Uruguay and Venezuela (suspended since 2016). Its associated members are Chile, Bolivia, Colombia, Ecuador, PerÃº, Surinam and Guyana. 
Sadly the report doesn't have data about Surinam and Guyana so they will not be considered here.

```python
mercosur=report_2021[(report_2021['Country name']=='Argentina') | (report_2021['Country name']=='Uruguay') | (report_2021['Country name']=='Brazil') | (report_2021['Country name']=='Paraguay') | (report_2021['Country name']=='Venezuela') | (report_2021['Country name']=='Chile') | (report_2021['Country name']=='Bolivia') | (report_2021['Country name']=='Colombia') | (report_2021['Country name']=='Peru') | (report_2021['Country name']=='Ecuador')]
mercosur
```

```python
plt.figure(figsize=(25, 8))
plt.title('Comparison of Ladder Scores in Mercosur countries', fontsize=24)
sns.barplot(data=mercosur, x='Country name', y='Ladder score')
```
